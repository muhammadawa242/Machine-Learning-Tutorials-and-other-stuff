# Stack Overflow

```python
filt = (df['Country'].str.contains('Pak', na=False)) & df['CompFreq'].str.contains('Yearly', na=False)
df.loc[filt,['CompTotal','CompFreq','Currency']].sort_values(by=['CompTotal'],ascending=[False])
```

```python
filt = df['Currency'].str.contains('USD', na=False)
df.loc[filt,['Employment','EdLevel','YearsCode','YearsCodePro','DevType','CompTotal','CompFreq','LanguageHaveWorkedWith','Ethnicity']].nlargest(100,'CompTotal')
```
